```python
import scipy.io
import numpy as np
import soundfile as sf
import librosa

MAT_PATH = "data/song21_Imputed.mat"
AUDIO_PATH = "data/songs/song21_16k.wav"

CHUNK_SEC = 3.5
EEG_FS = 125
AUDIO_FS = 16000

mat = scipy.io.loadmat(MAT_PATH)
data = mat["data21"]   # (125, 34795, 20)

print("data21 shape:", data.shape)

subj_idx = 0
eeg = data[:, :, subj_idx]   # (125, 34795)
print("single subject EEG:", eeg.shape)

audio, sr = sf.read(AUDIO_PATH)

if audio.ndim == 2:
    audio = audio.mean(axis=1)

if sr != AUDIO_FS:
    audio = librosa.resample(audio, orig_sr=sr, target_sr=AUDIO_FS)
    print(f"Resampled from {sr} Hz to {AUDIO_FS} Hz")
else:
    print(f"Audio already at {sr} Hz, no resampling needed")

print("audio shape:", audio.shape)

eeg_duration = eeg.shape[1] / EEG_FS
audio_duration = len(audio) / AUDIO_FS

print("EEG duration:", eeg_duration)
print("Audio duration:", audio_duration)

eeg_chunk_len = int(CHUNK_SEC * EEG_FS)      # 437
audio_chunk_len = int(CHUNK_SEC * AUDIO_FS)  # 56000

n_chunks_eeg = eeg.shape[1] // eeg_chunk_len
n_chunks_audio = len(audio) // audio_chunk_len
n_chunks = min(n_chunks_eeg, n_chunks_audio)

print("EEG chunk len:", eeg_chunk_len)
print("Audio chunk len:", audio_chunk_len)
print("n_chunks:", n_chunks)
```
